## Introduction

One of the hardest concepts in an introductory calculus class is understanding the difference between convergence of an infinite **sequence** and an infinite **series**. Let's briefly recall what these two objects are.

**Definition (Convergence of an Infinite Sequence)**

*An **infinite sequence** is an ordered collection of terms denoted by $\{a_n\}_{n=1}^\infty$, where $a_n$ can be defined either recursively or explicitly. For example the Fibonacci sequence $\{1,1,2,3,5,8,13,\ldots\}$ can be either recursively defined using $a_n = a_{n-1}+a_{n-2}$ or explicitly defined using $$a_n = \frac{\left(1+\sqrt 5\right)^n-\left(1-\sqrt 5\right)^n}{2^n\sqrt5}$$

An infinite sequence is said to **converge** if the limit $\lim_{n\to\infty} a_n$ exsits.*

For example, the Fibonacci sequence does not converge, and neither does the sequence $\{(-1)^n\} = \{-1,1,-1,1,\ldots\}$, but the Harmonic sequence $\left\{\frac1n\right\}$ does since $\lim_{n\to\infty}\frac1n = 0$. 

**Definition (Convergence of an Infinite Series)**

*An **infinite series** is the infinite sum $\sum_{i=1}^\infty a_i$. An infinite series is said to **converge** if the sequence made up of the partial sums $s_k = \sum_{n=1}^k a_n$ converges. *

For example consider the Harmonic sequence
$$\left\{\frac1n\right\}_{n=1}^\infty = \left\{1,\frac12,\frac13,\frac14,\ldots\right\}.$$
Then the first few partial sums are
$$
s_1 = a_1 = 1\quad s_2 = a_1+a_2 = 1+\frac12 = \frac 32\quad s_3 = a_1+a_2+a_3 = 1+\frac12+\frac13 = \frac{11}6
$$
and so on. Using these partial sums we can construct the sequence of partial sums 
$$
\left\{1,\frac32,\frac{11}6,\frac{25}{12},\frac{137}{60},\frac{49}{20}\ldots \right\} \approx\{1,1.5,1.83,2.08,2.28,2.45,\ldots\}.
$$
The question we want to ask is: does this **new** sequence converge? It is important to keep in mind that the answer is **not immediately obvious** from the computations above (and, in fact, turns out to be "no!"). The confusing point is that just because a **sequence** $\{a_n\}$ converges does not necessarily mean that the corresponding **series** $\sum_{n=1}^\infty a_n$ converges. 

This fact is closely related to the fact that just because a **function** $f(x)$ converges as $x\to\infty$ does not necessarily mean that the corresponding integral $\lim_{b\to\infty}\int_a^b f(x)dx$ converges. In fact, infinite sequences can be thought of as "discrete functions" and infinite series can be thought of as their "discrete integrals."

The purpose of this Notebook is to investigate the relationship between the convergence of a sequence and the convergence of a series.

## The Divergence Test
In the study of infinite series student learn a whole bunch of tests for convergence starting with one known as the Divergence Test which says

**Theorem (Divergence Test)**

*Let $\{a_n\}$ be an infinite sequence that does not converge to zero. Then the corresponding series $\sum_{n=1}^\infty a_n$ does not converge either.*

For example, the series $\sum_{n=1}^\infty (-1)^n$ does not converge since $\lim_{n\to\infty} (-1)^n$ does not exist. Similarly, $\sum_{n=1}^\infty n$ does not converge since $\lim_{n\to\infty} n$ does not exist. Furthermore, $\sum_{i=n}^\infty (1-\frac1n)$ does not convergence since $\lim_{n\to\infty} (1-\frac1n) = 1\neq 0$. Try these examples in the simulation above. This is related to the fact that the following integrals do not converge: $\int_1^\infty sin(x)dx$, $\int_1^\infty xdx$, and $\int_1^\infty \left(1-\frac1x\right)dx$. Try these examples in the simulation above. 

Students often misapply the Divergence Test by taking out the "nots" in the Theorem. In other words, they like to claim "Let $\{a_n\}$ be an infinite sequence that **does** converge to zero. Then the corresponding series $\sum_{i=1}^n a_n$ converges as well." **This is not true** as the simulations above show (go back and reexamine $a_n = \frac1{\sqrt n}$) and is a very common logical fallacy known as "fallacy of the inverse" or "denying the antecedent." Here's an example of the fallacy of the inverse:

"If I live in California then I live in the USA. Therefore if I do not live in California then I do not not live in the USA."

The above statement is **clearly not true**. I could live in New York, Floria, Alabama, or any one of 49 other states that are still in the USA. Whenever you're doing logical manipulations I highly recommend that you apply the same manipulation to the California-USA example above to see if it still makes sense. For example, the **contrapositive** **IS** true:

"If I live in California then I live in the USA. Therefore, if I do not live in the USA then I do not live in California."

What makes this statement different from the inverse is that **in addition** to putting "not"s in front of each statement I also **switched** the order. The contrapositive of the Divergence Test is:

**Theorem (Contrapositive of Divergence Test)**

*Let $\sum_{i=1}^\infty a_n$ be an infinite series that does converge. Then $a_n$ converges to zero.*

Be very, very careful that you use the contrapositive and not the inverse fallacy!
